```python
input_features = input.select_dtypes(include='number')
```

```python
input_features
```

```python
Q1 = input_features.quantile(0.25)
Q3 = input_features.quantile(0.75)
IQR = Q3 - Q1
mask = ~((input_features < (Q1 - 1.5 * IQR)) | (input_features > (Q3 + 1.5 * IQR))).any(axis=1)
input = input[mask].reset_index(drop=True)
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(data=input.select_dtypes(include='number'))
plt.title("Boxplot for All Numerical Columns")
plt.show()
```
